```python
v_A = np.array([-0.4,0.3,-0.6])
v_B = np.array([-0.2,0.2,1])
v_C = np.array([0.2,0.1,-0.5])

X = ((v_A**2+v_B**2+v_C**2)@(v_A*(v_B+v_A*v_B)/v_C))*(np.absolute(v_A))+(np.absolute(v_B))
Y = v_A**2+v_B**2
Z = (v_A)
X = (v_A@v_A + v_B@v_B + v_C@v_C)
X1 = (v_A*(v_B+v_A*v_B)/v_C)
X2 = (np.linalg.norm(v_A+v_B+v_C))
Ans = X*X1*X2

np.array(Ans)
```
